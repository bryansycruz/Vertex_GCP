# ¿Qué son las incrustaciones?

Con el auge de los LLMs, ¿por qué se está volviendo importante para los ingenieros de TI y los ITDMs entender cómo funcionan?

En los sistemas de TI tradicionales, la mayoría de los datos se organizan como datos estructurados o tabulares, utilizando palabras clave simples, etiquetas y categorías en bases de datos y motores de búsqueda.

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/1.png)

En contraste, los servicios impulsados por IA organizan los datos en una estructura de datos simple conocida como "incrustaciones".

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/2.png)

Una vez entrenado con contenido específico como texto, imágenes o cualquier contenido, la IA crea un espacio llamado "espacio de incrustaciones", que es esencialmente un mapa del significado del contenido.

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/3.png)

La IA puede identificar la ubicación de cada contenido en el mapa, eso es lo que es una incrustación.

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/4.png)

Tomemos un ejemplo donde un texto discute películas, música y actores, con una distribución del 10%, 2% y 30%, respectivamente. En este caso, la IA puede crear una incrustación con tres valores: 0.1, 0.02 y 0.3, en un espacio tridimensional.

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/5.png)

La IA puede colocar contenido con significados similares cerca en el espacio.

Así es como Google organiza datos en varios servicios como Google Search, YouTube, Play y muchos otros, para proporcionar resultados de búsqueda y recomendaciones con contenido relevante.

Las incrustaciones también se pueden usar para representar diferentes tipos de cosas en los negocios, como productos, usuarios, actividades de usuarios, conversaciones, música y videos, señales de sensores IoT, y más.

La IA y las incrustaciones ahora están desempeñando un papel crucial en la creación de una nueva forma de interacción humano-computadora.

## Casos de uso empresarial de incrustaciones de texto con LLM

Con la API de incrustaciones, puedes aplicar la innovación de las incrustaciones, combinada con la capacidad de los LLM, a varias tareas de procesamiento de texto, tales como:

**Búsqueda semántica habilitada por LLM**: las incrustaciones de texto se pueden usar para representar tanto el significado como la intención de la consulta de un usuario y los documentos en el espacio de incrustaciones. Los documentos que tienen un significado similar a la intención de la consulta del usuario se encontrarán rápidamente con la tecnología de búsqueda vectorial. El modelo es capaz de generar incrustaciones de texto que capturan los matices sutiles de cada oración y párrafo en el documento.

**Clasificación de texto habilitada por LLM**: las incrustaciones de texto de LLM se pueden usar para la clasificación de texto con una comprensión profunda de diferentes contextos sin necesidad de entrenamiento o ajuste fino (lo que se llama aprendizaje de cero disparos). Esto no era posible con los modelos de lenguaje anteriores sin entrenamiento específico para la tarea.

**Recomendación habilitada por LLM**: la incrustación de texto se puede usar para sistemas de recomendación como una característica sólida para entrenar modelos de recomendación como el modelo Two-Tower. El modelo aprende la relación entre las incrustaciones de consulta y las candidatas, lo que resulta en una experiencia de usuario de próxima generación con recomendaciones de productos semánticos.

La agrupación, detección de anomalías, análisis de sentimientos y más habilitados por LLM también se pueden manejar con la comprensión semántica profunda a nivel de LLM.

### Preparación de datos

Usaremos [el conjunto de datos público de Stack Overflow](https://console.cloud.google.com/marketplace/product/stack-exchange/stack-overflow) alojado en la tabla de BigQuery `bigquery-public-data.stackoverflow.posts_questions`. Este es un conjunto de datos muy grande con 23 millones de filas que no cabe en memoria. Vamos a limitarlo a 1000 filas para este tutorial.

```python
# load the BQ Table into a Pandas DataFrame
from google.cloud import bigquery

QUESTIONS_SIZE = 1000

bq_client = bigquery.Client(project=PROJECT_ID)
QUERY_TEMPLATE = """
        SELECT distinct q.id, q.title
        FROM (SELECT * FROM `bigquery-public-data.stackoverflow.posts_questions`
        where Score > 0 ORDER BY View_Count desc) AS q
        LIMIT {limit} ;
        """
query = QUERY_TEMPLATE.format(limit=QUESTIONS_SIZE)
query_job = bq_client.query(query)
rows = query_job.result()
df = rows.to_dataframe()

# examine the data
df.head()
```

### Call the API to generate embeddings

With the Stack Overflow dataset, we will use the `title` column (the question title) and generate embedding for it with Embeddings for Text API. The API is available under the [vertexai](https://cloud.google.com/python/docs/reference/aiplatform/latest/vertexai) package of the SDK.

You may see some warning messages from the TensorFlow library but you can ignore them.

Desde el paquete, importa `TextEmbeddingModel` y obtiene un modelo.

```python
# init the vertexai package
from google import genai
from google.genai import types

EMBEDDING_MODEL = "gemini-embedding-001"
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
```

En este tutorial usaremos el modelo `gemini-embedding-001` para obtener incrustaciones de texto. Consulta los [modelos compatibles](https://cloud.google.com/vertex-ai/docs/generative-ai/embeddings/get-text-embeddings#supported_models) en la documentación para ver la lista completa.

Una vez que tengas el modelo, puedes llamar a su función `get_embeddings` para obtener incrustaciones. Puedes pasar hasta 5 textos a la vez en una llamada. Hay una advertencia: por defecto la API de incrustaciones tiene una cuota de "solicitudes por minuto" establecida en 60 para proyectos nuevos y 600 para proyectos con historial de uso (consulta las [Cuotas y límites](https://cloud.google.com/vertex-ai/docs/quotas#request_quotas) para verificar el valor más reciente de la cuota para `base_model:textembedding-gecko`). Por ello, en lugar de llamar a la función directamente, conviene definir un envoltorio que limite a menos de 10 llamadas por segundo y pase 5 textos cada vez.

```python
# get embeddings for a list of texts
BATCH_SIZE = 5


def get_embeddings_wrapper(texts: list[str]) -> list[list[float]]:
    embeddings: list[list[float]] = []
    for i in tqdm.tqdm(range(0, len(texts), BATCH_SIZE)):
        time.sleep(1)  # to avoid the quota error
        response = client.models.embed_content(
            model=EMBEDDING_MODEL, contents=texts[i : i + BATCH_SIZE]
        )
        embeddings = embeddings + [e.values for e in response.embeddings]
    return embeddings
```

El siguiente código obtendrá la incrustación de los títulos de las preguntas y las añadirá como una nueva columna `embedding` al DataFrame. Esto tomará unos minutos.

```python
# get embeddings for the question titles and add them as "embedding" column
df = df.assign(embedding=get_embeddings_wrapper(list(df.title)))
df.head()
```

## Observa las similitudes de las incrustaciones

Veamos cómo se organizan estas incrustaciones en el espacio de incrustaciones según su significado calculando rápidamente las similitudes entre ellas y ordenándolas.

Como las incrustaciones son vectores, puedes calcular la similitud entre dos incrustaciones usando una de las métricas populares como las siguientes:

![](https://storage.googleapis.com/github-repo/img/embeddings/textemb-vs-notebook/8.png)

¿Qué métrica usar? Depende de cómo se haya entrenado cada modelo. En el caso del modelo `gemini-embedding-001`, debemos usar el producto interno (dot product).

En el siguiente código, se selecciona una pregunta al azar y se usa la función `np.dot` de numpy para calcular las similitudes entre esa pregunta y las demás.

```python
# pick one of them as a key question
key = random.randint(0, len(df))

# calc dot product between the key and other questions
embs = np.array(df.embedding.to_list())
similarities = np.dot(embs[key], embs.T)

# print similarities for the first 5 questions
similarities[:5]
```

Finalmente, ordena las preguntas por similitud e imprime la lista.

```python
# print the question
print(f"Key question: {df.title[key]}\n")

# sort and print the questions by similarities
sorted_questions = sorted(
    zip(df.title, similarities), key=lambda x: x[1], reverse=True
)[:20]
for i, (question, similarity) in enumerate(sorted_questions):
    print(f"{similarity:.4f} {question}")
```

# Encontrar incrustaciones rápidamente con Vertex AI Vector Search

Como explicamos arriba, puedes encontrar incrustaciones similares calculando la distancia o similitud entre las incrustaciones.

Pero esto no es fácil cuando tienes millones o miles de millones de incrustaciones. Por ejemplo, si tienes 1 millón de incrustaciones con 768 dimensiones, tendrías que repetir los cálculos de distancia 1 millón x 768 veces. Esto tomaría segundos — demasiado lento.

Por eso los investigadores han estudiado una técnica llamada [Approximate Nearest Neighbor (ANN)](https://en.wikipedia.org/wiki/Nearest_neighbor_search) para búsquedas más rápidas. ANN usa "cuantización vectorial" para separar el espacio en múltiples subespacios con una estructura tipo árbol. Esto es similar al índice en bases de datos relacionales para mejorar el rendimiento de las consultas, permitiendo búsquedas muy rápidas y escalables con miles de millones de incrustaciones.

Con el auge de los LLMs, ANN se ha popularizado rápidamente bajo el nombre de tecnología de Búsqueda Vectorial.

### Create an Index

Now it's ready to load the embeddings to Vector Search. Its APIs are available under the [aiplatform](https://cloud.google.com/python/docs/reference/aiplatform/latest/google.cloud.aiplatform) package of the SDK.

```python
# init the aiplatform package
from google.cloud import aiplatform

aiplatform.init(project=PROJECT_ID, location=LOCATION)
```

Create an [MatchingEngineIndex](https://cloud.google.com/python/docs/reference/aiplatform/latest/google.cloud.aiplatform.MatchingEngineIndex) with its `create_tree_ah_index` function (Matching Engine is the previous name of Vector Search).

```python
# create index
from google.cloud import aiplatform

PROJECT_ID = "qwiklabs-gcp-04-b9fc3d810780"
LOCATION = "us-east4"
BUCKET_URI = f"gs://{PROJECT_ID}-embvs-tutorial-{UID}"

aiplatform.init(project=PROJECT_ID, location=LOCATION)

my_index = aiplatform.MatchingEngineIndex.create_tree_ah_index(
    display_name="embvs-tutorial-index",
    contents_delta_uri=BUCKET_URI,
    dimensions=3072,
    approximate_neighbors_count=20,
    distance_measure_type="DOT_PRODUCT_DISTANCE",
    leaf_node_embedding_count=100,        # optional
    leaf_nodes_to_search_percent=10       # optional
)
```

By calling the `create_tree_ah_index` function, it starts building an Index. This will take under a few minutes if the dataset is small, otherwise about 50 minutes or more depending on the size of the dataset. You can check status of the index creation on [the Vector Search Console > INDEXES tab](https://console.cloud.google.com/vertex-ai/matching-engine/indexes).

![](https://storage.googleapis.com/github-repo/img/embeddings/vs-quickstart/creating-index.png)

#### The parameters for creating index

- `contents_delta_uri`: The URI of Cloud Storage directory where you stored the embedding JSON files
- `dimensions`: Dimension size of each embedding. In this case, it is 3072 as we are using the embeddings from the Text Embeddings API.
- `approximate_neighbors_count`: how many similar items we want to retrieve in typical cases
- `distance_measure_type`: what metrics to measure distance/similarity between embeddings. In this case it's `DOT_PRODUCT_DISTANCE`

See [the document](https://cloud.google.com/vertex-ai/docs/vector-search/create-manage-index) for more details on creating Index and the parameters.

#### Batch Update or Streaming Update?
There are two types of index: Index for *Batch Update* (used in this tutorial) and Index for *Streaming Updates*. The Batch Update index can be updated with a batch process whereas the Streaming Update index can be updated in real-time. The latter one is more suited for use cases where you want to add or update each embeddings in the index more often, and crucial to serve with the latest embeddings, such as e-commerce product search.


### Crear Endpoint de Índice y desplegar el Índice

Para usar el Índice, necesitas crear un [Index Endpoint](https://cloud.google.com/vertex-ai/docs/vector-search/deploy-index-public). Funciona como una instancia de servidor que acepta solicitudes de consulta para tu Índice.

```python
# create IndexEndpoint
my_index_endpoint = aiplatform.MatchingEngineIndexEndpoint.create(
    display_name=f"embvs-tutorial-index-endpoint-{UID}",
    public_endpoint_enabled=True,
)
```

Este tutorial utiliza un [Endpoint público](https://cloud.google.com/vertex-ai/docs/vector-search/setup/setup#choose-endpoint) y no admite [Virtual Private Cloud (VPC)](https://cloud.google.com/vpc/docs/private-services-access). A menos que tengas un requisito específico para VPC, recomendamos usar un Endpoint público. A pesar del término "público" en su nombre, no implica acceso abierto a Internet: funciona como otros endpoints en los servicios de Vertex AI y está asegurado por IAM de forma predeterminada. Sin permisos IAM explícitos, nadie puede acceder al endpoint.

With the Index Endpoint, deploy the Index by specifying an unique deployed index ID.

```python
DEPLOYED_INDEX_ID = f"embvs_tutorial_deployed_{UID}"
```

```python
# deploy the Index to the Index Endpoint
my_index_endpoint.deploy_index(
    index=my_index,
    deployed_index_id=DEPLOYED_INDEX_ID,
    machine_type="e2-standard-16",
    min_replica_count=1,
    max_replica_count=1  
)
```

Si es la primera vez que despliegas un Índice en un Index Endpoint, tardará alrededor de 25 minutos en construir e iniciar automáticamente el backend. Después del primer despliegue, tardará segundos. Para ver el estado del despliegue del índice, abre [la Consola de Vector Search > pestaña INDEX ENDPOINTS](https://console.cloud.google.com/vertex-ai/matching-engine/index-endpoints) y haz clic en el Index Endpoint.

# Resumen

## Fundamentar las salidas del LLM con Vertex AI Vector Search

Como hemos visto, combinando la API de Incrustaciones y Vector Search, puedes usar las incrustaciones para "fundamentar" las salidas de un LLM con datos empresariales reales y baja latencia.

Por ejemplo, si un usuario hace una pregunta, la API de Incrustaciones puede convertirla en una incrustación y realizar una consulta en Vector Search para encontrar incrustaciones similares en su índice. Esas incrustaciones representan los datos reales de negocio en las bases de datos. Como solo estamos recuperando datos empresariales y no generando textos artificiales, no existe riesgo de alucinaciones en el resultado.

![](https://storage.googleapis.com/gweb-cloudblog-publish/original_images/10._grounding.png)

### La diferencia entre preguntas y respuestas

En este tutorial usamos el conjunto de datos de Stack Overflow. Hay una razón: el conjunto contiene muchos pares de **preguntas y respuestas**, por lo que puedes encontrar preguntas similares a la tuya para hallar respuestas.

En muchos casos de negocio, la semántica (significado) de preguntas y respuestas puede ser diferente. Además, puede que quieras añadir variedad o elementos personalizados a los resultados, como en la búsqueda de productos en e‑commerce.

En estos casos la simple búsqueda semántica no funciona bien. Es más parecido a un problema de recomendación donde podrías querer entrenar un modelo (por ejemplo, Two‑Tower) para aprender la relación entre el espacio de incrustaciones de la consulta y el de la respuesta. Además, muchos sistemas de producción añaden una fase de reordenamiento (reranking) tras la búsqueda semántica para mejorar la calidad. Consulta [Scaling deep retrieval with TensorFlow Recommenders and Vertex AI Matching Engine](https://cloud.google.com/blog/products/ai-machine-learning/scaling-deep-retrieval-tensorflow-two-towers-architecture) para más información.

### Híbrido de búsqueda semántica + por palabras clave

Otro reto típico en producción es soportar búsqueda por palabras clave combinada con búsqueda semántica. Por ejemplo, en e‑commerce puede que quieras permitir que los usuarios busquen por nombre de producto o número de modelo. Como los LLM no memorizan necesariamente esos nombres o números, la búsqueda semántica no puede manejar esas funcionalidades por sí sola.

[Vertex AI Search](https://cloud.google.com/blog/products/ai-machine-learning/vertex-ai-search-and-conversation-is-now-generally-available) es otro producto a considerar para esos requisitos. Mientras Vector Search proporciona una búsqueda semántica simple, Search ofrece una solución integrada que combina búsqueda semántica, búsqueda por palabras clave, reranking y filtrado, lista para usar.

### ¿Y qué hay de Retrieval Augmented Generation (RAG)?

En este tutorial hemos visto la combinación simple de incrustaciones y búsqueda vectorial. A partir de aquí puedes extender el diseño a [Retrieval Augmented Generation (RAG)](https://www.google.com/search?q=Retrieval+Augmented+Generation+(RAG)&oq=Retrieval+Augmented+Generation+(RAG)).

RAG es una arquitectura popular para implementar fundamentación con LLMs y una interfaz de chat basada en texto. La idea es usar la UI de chat como frontend para la recuperación de documentos mediante búsqueda vectorial y la posterior summarización de los resultados.

![](https://storage.googleapis.com/gweb-cloudblog-publish/images/Figure-7-Ask_Your_Documents_Flow.max-529x434.png)

Ambas soluciones tienen pros y contras.

| | Embeddings + búsqueda vectorial | RAG |
|---|---|---|
| Diseño | simple | complejo |
| UI | Interfaz de búsqueda por texto | UI de chat por texto |
| Resumen de resultados | No | Sí |
| Multi‑turn (contexto) | No | Sí |
| Latencia | milisegundos | segundos |
| Coste | menor | mayor |
| Alucinaciones | Sin riesgo | Algo de riesgo |

El patrón Incrustaciones + búsqueda vectorial que hemos visto ofrece búsqueda semántica simple, rápida y de bajo coste. RAG añade experiencia tipo chat contextualizada y resumen de resultados, pero también incrementa el coste y el riesgo de alucinaciones.

Para aprender cómo construir una solución RAG, consulta [Building Generative AI applications made easy with Vertex AI PaLM API and LangChain](https://cloud.google.com/blog/products/ai-machine-learning/generative-ai-applications-with-vertex-ai-palm-2-models-and-langchain).
